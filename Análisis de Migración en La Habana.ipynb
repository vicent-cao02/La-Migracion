### Según la Revista Novedades en Población, la migración es la tercera variable que ejerce un efecto en la estructura demográfica de la población, asi como en sus cantidades de efectivos, su contribución al cambio de la población puede ser pequeña pero esta varía en la medida en que los análisis toman unidades administrativas menores. La migración interna y externa en La Habana ha experimentado cambios significativos a lo largo de las últimas décadas influenciado por factores políticos y sociales.Apoyándose en los datos que aporta el Anuario Estadístico de Cuba se observa que en el año 2022 se registraron 19 279 personas que migraron hacia y desde municipios de la capital. Las mayores entradas se produjeron en los municipios de 10 de Octubre y Boyeros, no siendo así en el municipio de La Habana Vieja donde se registró una entrada de 286 personas, esto debido a que la infraestructura y los servicios pueden no ser tan accesibles o eficientes como en otras áreas de la ciudad lo que podría disuadir a los migrantes a establecerse aquí. Además, se destacan las mayores corrientes migratorias hacia la urbe capitalina provenientes de las provincias de Holguín y Granma, siguiéndole Santiago de Cuba.

```python
#Gráfico de línea para representar la migración interna y externa en la Habana 
import pandas as pd 
df = pd.read_csv('Data/Saldos_migratorios_y_tasa_de_saldo_migratorio_total_por_provincias.csv')

import matplotlib.pyplot as plt
df.plot(x='Annos', y=['Saldo Migratorio interno', 'Saldo Migratorio externo'], kind="line", figsize=(15,7))
plt.show()
```

### Todos los municipios registraron saldos migratorios positivos. El mayor porciento de todos los movimientos internos en el país dirigidos a municipios de La Habana se produjo entre sus propios territorios, lo que indica una migración intrametropolitana significativa. En los últimos cinco años en términos de género las tasas de saldo migratorio son superiores en hombres que en mujeres.

### Por otra parte la migración interna en Cuba ha tenido un impacto económico significativo, afectando la distribución de la población, el mercado laboral, y potencialmente el sector agrícola. Las políticas migratorias implementadas, como el Decreto-Ley 217, buscaba controlar la migración interna a La Habana, especialmente en respuesta a la creciente demanda de vivienda y empleo en la capital, intentando controlar este fenómeno, pero los efectos económicos y sociales de la migración interna continúan siendo una preocupación importante para el país. Durante el Período Especial, marcado por la desintegración del campo socialista y la reducción de inversiones en Cuba, se observaron efectos negativos económicos y sociales, incluyendo un aumento en la migración descontrolada hacia las mayores ciudades y la Ciudad de La Habana. La migración interna ha jugado un papel redistributivo en la población desde varias décadas atrás. Esto ha llevado a cambios en las prácticas laborales y en la estructura del mercado laboral razonable.
